## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

with sns.axes_style('whitegrid'):

  grafico_price = sns.boxplot(x=data["price"])
  grafico_price.set(title='Recorrência do Preço de Diamantes', xlabel='Valor');
```

**Insight do gráfico 1**:
- A partir de cerca do valor de 11800, o preço estão mais distantes um de outro (outliers).
- O preço mais comum de diamantes está na faixa de, aproximadamente, 800 e 5200.
- Diamantes mais baratos são mais comuns, enquanto os mais caros são mais raros.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 2

diamonds = data[["price", "carat", "clarity"]]
diamonds.head()
```

```
with sns.axes_style('whitegrid'):

  grafico_diamonds = sns.scatterplot(data=diamonds, x="carat", y="price", hue="clarity", palette="bright")
  grafico_diamonds.set(title='Correlação entre o do Preço do Diamante com seu Peso Agrupados por sua Transparêcia', xlabel='Peso', ylabel='Preço');
  grafico_diamonds.get_legend().set_title("Claridade");
```

**Insight do gráfico 2**:
- Majoritariamente, ao aumentar o peso (carat), o preço também aumenta.
- A proporção de aumento do preço em relação peso, aumenta de acordo com a claridade, sendo a menor proporção IF e a maior proporção I1.
- Praticamente todas as claridades apresentam crescimento parecido, com exeção da I1, que está mais distante da curva de crescimento das outras claridades.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3

with sns.axes_style('whitegrid'):

  grafico_cars = sns.histplot(data=data, x="weight", kde=True)
  grafico_cars.set(title='Distribuição Contínua aproximada do Peso de Carros', xlabel='Peso', ylabel='Contagem');
```

**Insight do gráfico 3**:
- O peso mais comum para carros é de, aproximadamente, 2200.
- O peso mais incomum para carros é de, aproximadamente, 5200.
- Há um crescimento no peso entre, aproximadamente, 1600-2200 e logo depois há uma queda contínua e praticamente constante até, aproximadamente, 5200.
